```python
import pandas as pd 
import numpy as np
k= np.random.randint(1,30, size = (5,3))
df = pd.DataFrame(k, columns = ["met1","met2","met3"])
df
```

```python
df2= df +100
df2
```

```python
pd.concat([df,df2])  #JOIN İŞLEMİ concat fonksiyonu ile yapılır
```

```python
pd.concat([df,df2], ignore_index=True)  #Yukarıdaki index hatasını düzeltmek için 'ignore_index=true' fonksiyonunu kullanırız
#abcde gitti. yerine 0 dan 9 a kadar index no ları geldi.
```

```python
#EGER COLUMS LARDA FARKLILIK VARSA JOIN İSLEMİ BASSRILI OLABİLİR AMA AYNI CESİT DEGERİ TASIMAYANLAR HATALI CIKTI VERİR
#AŞAGIDAKİ İŞLEMLERDE  GÖZLEMLENEN HATALAR VE DÜELTME YÖNTEMLERİNE DİKKAT EDİNİZ.
```

```python
df2.columns = ["met1","met2", "MAT1"]
df2
```

```python
pd.concat([df,df2]) # işte tamda burada aşagıdaki gibi bir hata NaN hatası var bunun düzeltilmesi lazım..'inner joın'
```

```python
pd.concat([df,df2] , ignore_index=True, join = "inner") # HER İKİ dATAFRAMDE Kİ KESİMLERİ ALDIKK.. 
```

```python
pd.concat([df,df2] , ignore_index=True, join_axes = [df.columns]) # burada df nin columlarını sabit tututp df2 nin farklı olaan
  #columnunu df nin devamına yazmıs olduk
```
